```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time
pd.set_option('display.max_columns',30)
pd.set_option('display.width', 1000)
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
import requests
from bs4 import BeautifulSoup
import json
```

```python
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
driver.maximize_window()
driver.get('https://desa.pl/pl/wyniki-aukcji-dziel-sztuki/')
```

```python
#rozwijanie listy
while(1<2):
    try:
        driver.find_element(By.XPATH, "/html/body/div[3]/div[3]/div[3]/form/div[1]").click()
        time.sleep(2)
    except:
        break
```

```python
#linki aukcji desa
aukcje=driver.find_elements(By.CSS_SELECTOR, 'a.box-object-row__link.link-underlined')
linki_aukcji=[]
for el in aukcje:
    if  el.get_attribute('href')[:11]!='https://bid' and el.get_attribute('href')[:15]!='https://www.bid':
        linki_aukcji.append(el.get_attribute('href'))
export=pd.DataFrame(linki_aukcji)
export.to_csv('C:/Users/mikol/Documents/python-projects/linki_aukcji.csv', index = False, header=False)
```

```python
linki_aukcji=pd.read_csv('C:/Users/mikol/Documents/python-projects/linki_aukcji.csv', header=None)
linki_aukcji=linki_aukcji[0].values.tolist()
linki_aukcji
```

```python
#linki aukcji desa_bid
link_bid=[]
link_desa=[]
for i in linki_aukcji:
    print(i)
    page = requests.get(i)
    soup = BeautifulSoup(page.content, 'html.parser')
    skrypt = soup.find_all('script', type='application/ld+json')[0].text.strip().replace('\r\n',', ').replace('','')
    data = json.loads(skrypt)
    try:
        link_bid.append(data['location'][0]['url'])
    except:
        continue
    link_desa.append(i)
        
export=pd.DataFrame({'link_desa':link_desa, 'link_bid':link_bid})
export.to_csv('C:/Users/mikol/Documents/python-projects/desa_bid.csv', index = False, header=True)
```

```python
#liniki przedmiotow desa
link_przedmiotu=[]
link_aukcji=[]

for i in linki_aukcji:
    driver.get(i)
    time.sleep(9)
    try:
        driver.find_element(By.XPATH, "/html/body/div[1]/div[7]/div[2]/div[3]/div/a[2]").click()
        time.sleep(10)
    except:
        try:
            driver.find_element(By.XPATH, "/html/body/div[1]/div[8]/div[2]/div[3]/div/a[2]").click()
            time.sleep(10)
        except:
            try:
                driver.find_element(By.XPATH, "/html/body/div[1]/div[6]/div[2]/div[3]/div/a[2]").click()
                time.sleep(10)
            except:
                pass
    aukcje=driver.find_elements(By.CSS_SELECTOR, 'a.box-object-row__link.link-underlined')
    for el in aukcje:
        link_przedmiotu.append(el.get_attribute('href'))
    link_aukcji.extend(len(aukcje)*[i])
    time.sleep(1)
linki_przedmiotow=pd.DataFrame({'przedmiot':link_przedmiotu, 'aukcja':link_aukcji})
linki_przedmiotow.to_csv('C:/Users/mikol/Documents/python-projects/linki_przedmiotow.csv', index = False, header=True)
```

```python
#linki przedmiotów bid.desa
przypadki=[]
link_aukcji=[]

for i in linki_aukcji:
    driver.get(i+'?limit=1000')
    time.sleep(8)
    przedmioty=driver.find_elements(By.CSS_SELECTOR, 'a.block-link.ng-binding')
    for el in przedmioty:
        przypadki.append(el.get_attribute('href'))
    link_aukcji.extend(len(przedmioty)*[i])
linki_przedmiotow=pd.DataFrame({'przedmiot':przypadki, 'aukcja':link_aukcji})
linki_przedmiotow.to_csv('C:/Users/mikol/Documents/python-projects/linki_przedmiotow.csv', index = False, header=True)
```

```python
##zmienne aukcji
data=[]
tryb=[]
katalog=[]
film_promocyjny=[]
nazwa=[]
kod_aukcji=[]
opis=[]
lokalizacja=[]
departament=[]
#liczba_obiektow_wystawionych=[]
#liczba_obiektow_sprzedanych=[]
link_aukcji=[]
wystawa=[]

for i in linki_aukcji:
    page = requests.get(i)
    soup = BeautifulSoup(page.content, 'html.parser')
    #zmienne
    #pominięcie aukcji dla pełnoletnich
    if soup.find(class_="prompt-container") is not None:
        continue
    try:
        tryb.append(soup.find(class_="typography__type d-none d-lg-block").get_text().strip())
    except:
        tryb.append('NaN')
    try:
        data.append(soup.find(class_="typography__start-date d-none d-lg-block").get_text().strip())
    except:
         data.append('NaN')
    try:
        nazwa.append(soup.find(class_="typography__title").get_text().strip())
    except:
        try:
            nazwa.append(soup.find(class_="page-title").get_text().strip())
        except:
            nazwa.append('NaN')
    try:
        lokalizacja.append(soup.find(class_="map-pin-wrapper").get_text().strip().replace("\r\n", ", "))
    except:
        lokalizacja.append('NaN')
    try:
        departament.append(soup.find(class_="auction-department__value").get_text().strip())
    except:
        departament.append('NaN')
    try:
        wystawa.append(soup.find(class_="typography__dates d-none d-lg-block").get_text().strip()[9:])
    except:
        wystawa.append('NaN')
    katalog.append(1 if soup.find(class_="common-box__img bg-contain") is not None else 0)
    film_promocyjny.append(1 if soup.find(class_="common-box__img bg-cover") is not None else 0)
    opis.append(1 if soup.find(class_="description-txt-section") is not None else 0)
    try:
        kod_aukcji.append(soup.find(class_="auction-code__value").get_text().strip())
    except:
        kod_aukcji.append('NaN')
    link_aukcji.append(i)
   
dane_aukcji=pd.DataFrame({'nazwa':nazwa, 'kod_aukcji':kod_aukcji, 'departament':departament, 'lokalizacja':lokalizacja, 'tryb':tryb, 'data':data, 'wystawa':wystawa, 'katalog':katalog, 'film_promocyjny':film_promocyjny, 'opis':opis, 'link_aukcji':link_aukcji})
dane_aukcji.to_csv('C:/Users/mikol/Documents/python-projects/dane_aukcji.csv', index = False, header=True)

```

```python
#zmienne przedmiotow desa
autor=[]
#plec=[]
cena=[]
urodzony=[]
rok_namalowania=[]
tytul=[]
opis=[]
link=[]
for i in linki_przedmiotow['przedmiot']:
        driver.get(i)
        time.sleep(10)
        cena.append(int(driver.find_element(By.CSS_SELECTOR, 'span.desa-object__price-value').text.replace(" ","")[:-2]))
        autor.append(driver.find_element(By.CSS_SELECTOR, 'div.default-info__title').find_elements(By.XPATH, "./child::*")[0].text.split('(')[0][:-1])
        urodzony.append(driver.find_element(By.CSS_SELECTOR, 'div.default-info__title').find_elements(By.XPATH, "./child::*")[0].text.split('(')[1][:-1])
        rok_namalowania.append(int(driver.find_element(By.CSS_SELECTOR, 'div.default-info__title').find_elements(By.XPATH, "./child::*")[1].text[-4:]))
        tytul.append(driver.find_element(By.CSS_SELECTOR, 'div.default-info__title').find_elements(By.XPATH, "./child::*")[1].text[:-6])
        opis.append(driver.find_element(By.CSS_SELECTOR, 'div.default-info__desc').text.split('\nID: ')[0].replace('\n', ', '))
        link.append(i)
        
zmienne_przedmiotow=pd.DataFrame({'cena':cena, 'autor':autor, 'tytul':tytul, 'urodzony':urodzony, 'rok_namalowania':rok_namalowania, 'opis':opis, 'link':link})
zmienne_przedmiotow.to_csv('C:/Users/mikol/Documents/python-projects/zmienne_przedmiotow.csv', index = False, header=True)
```

```python
#zmienne przedmiotow bid.desa
cena=[]
wymiary=[]
link=[]
kategoria=[]
autor=[]
technika=[]
sygnatura=[]
tytul=[]
wystawiany=[]
opis=[]
literatura=[]
proweniencja=[]
estymacja=[]


#plec=[]
#rok_namalowania=[]
#rok_urodzenia=[]
#rok_smierci=[]

for i in linki_przedmiotow['przedmiot']:
        driver.get(i)
        time.sleep(5)
        try:
            tytul.append(driver.find_element(By.CSS_SELECTOR, 'div.lot-title.ng-binding').text)
        except:
            tytul.append('NaN')
        try:
            cena.append(int(driver.find_element(By.CSS_SELECTOR, 'div.sold-amount.amount.ng-binding').text.replace(" ","").replace(",","")[:-2]))
        except:
            cena.append('NaN')
        try:
            wymiary.append(driver.find_element(By.CSS_SELECTOR, 'div.description.ng-binding[ng-bind-html="lot.dimensions"]').text.replace(' ', '').replace('\n', ', '))
        except:
            wymiary.append('NaN')
        try:
            kategoria.append(driver.find_element(By.CSS_SELECTOR, 'a.description.category.ng-binding').text)
        except:
            kategoria.append('NaN')
        try:
            autor.append(driver.find_element(By.CSS_SELECTOR, 'a.description.artist.ng-binding').text)
        except:
            autor.append('NaN')
        try:
            technika.append(driver.find_element(By.CSS_SELECTOR, 'div.description.ng-binding[ng-bind-html="lot.medium"]').text)
        except:
            technika.append('NaN')
        try:
            sygnatura.append(driver.find_element(By.CSS_SELECTOR, 'div.description.ng-binding[ng-bind-html="lot.signature"]').text)
        except:
            sygnatura.append('NaN')
        try:
            wystawiany.append(driver.find_element(By.CSS_SELECTOR, 'div.description.ng-binding.ng-scope[ng-bind-html="lot.exhibited"]').text.replace(';', ',').replace('\n', ', '))
        except:
            wystawiany.append('NaN')
        try:
            opis.append(driver.find_element(By.CSS_SELECTOR, 'div.description.ng-binding[ng-bind-html="lot.description"]').text.split('Opłaty')[0].replace("\r\n", " ").replace(';', ' ').replace('\n', ' '))
        except:
            opis.append('NaN')
        try:
            proweniencja.append(driver.find_element(By.CSS_SELECTOR, 'div.description.ng-binding.ng-scope[ng-bind-html="lot.provenance"]').text.replace(';', ',').replace('\n', ', '))
        except:
            proweniencja.append('NaN')
        try:
            literatura.append(driver.find_element(By.CSS_SELECTOR, 'div.description.ng-binding.ng-scope[ng-bind-html="lot.literature"]').text.replace(';', ',').replace('\n', ', '))
        except:
            literatura.append('NaN')
        try:
            estymacja.append(driver.find_element(By.CSS_SELECTOR, 'div.range.ng-binding.ng-scope[ng-if="!viewVars.lotCurrencies"]').text.replace(',', '').replace('zł', '').replace(' ',''))
        except:
            estymacja.append('NaN')
        link.append(i)

zmienne_przedmiotow=pd.DataFrame({'estymacja':estymacja, 'cena':cena, 'autor':autor, 'tytul':tytul, 'kategoria':kategoria, 'wymiary':wymiary, 'technika':technika, 'sygnatura':sygnatura, 'opis':opis, 'proweniencja':proweniencja, 'literatura':literatura, 'wystawiany':wystawiany, 'link':link})
zmienne_przedmiotow.to_csv('C:/Users/mikol/Documents/python-projects/zmienne_przedmiotow.csv', index = False, header=True)
```
